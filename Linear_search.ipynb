```python
import random
def create_grid(rows, cols, num_obstacles, x, goal_pos):
    grid = [[0 for _ in range(cols)] for _ in range(rows)]  # Initialize grid with open spaces (0)
    for _ in range(num_obstacles):
        while True:
            r, c = random.randint(0, rows-1), random.randint(0, cols-1)
            if (r, c) != x and (r, c) != goal_pos and grid[r][c] == 0:
                grid[r][c] = 1  # Set obstacle
                break
    grid[x[0]][x[1]] = 'C'
    grid[goal_pos[0]][goal_pos[1]] = 'G'
    return grid

def print_grid(grid):
    for row in grid:
        print(' '.join(str(cell) for cell in row))
    print()
def move_car(grid, x, goal_pos):
    rows, cols = len(grid), len(grid[0])
    while x != goal_pos:
        print(f"Car is at {x}. Moving...")
        print_grid(grid)

        next_move = None
        r, c = x
        if c + 1 < cols and grid[r][c + 1] == 0:  # Move right
            next_move = (r, c + 1)
        elif r + 1 < rows and grid[r + 1][c] == 0:  # Move down
            next_move = (r + 1, c)
        elif c - 1 >= 0 and grid[r][c - 1] == 0:  # Move left
            next_move = (r, c - 1)
        elif r - 1 >= 0 and grid[r - 1][c] == 0:  # Move up
            next_move = (r - 1, c)
        if next_move:
            grid[x[0]][x[1]] = 0  # Clear old position
            x = next_move
            grid[x[0]][x[1]] = 'C'  # Set new position
        else:
            print("No valid move found, car is stuck!")
            break

    if x == goal_pos:
        print("Car reached the goal!")
        print_grid(grid)

rows, cols = 5, 5
car_start = (0, 0)
goal = (4, 4)
num_obstacles = 5

grid = create_grid(rows, cols, num_obstacles, car_start, goal)


print("Initial Grid:")
print_grid(grid)

move_car(grid, car_start, goal)
```

```python
import pygame

# Initialize Pygame
pygame.init()

# Define constants
GRID_SIZE = 20  # 20x20 grid
CELL_SIZE = 25  # Each cell is 25x25 pixels
SCREEN_WIDTH = GRID_SIZE * CELL_SIZE
SCREEN_HEIGHT = GRID_SIZE * CELL_SIZE
FPS = 5 # Frames per second

# Colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)
GREEN = (0, 255, 0)
BLUE = (0, 0, 255)

# Game objects
car_pos = [0, 0]  # Starting position of the car
goal_pos = [GRID_SIZE - 2, GRID_SIZE - 5]  # Goal position
obstacles = []  # List to store obstacle positions

# Initialize Pygame window
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("Car Simulation with Obstacles")

# Draw grid lines
def draw_grid():
    for x in range(0, SCREEN_WIDTH, CELL_SIZE):
        for y in range(0, SCREEN_HEIGHT, CELL_SIZE):
            rect = pygame.Rect(x, y, CELL_SIZE, CELL_SIZE)
            pygame.draw.rect(screen, WHITE, rect, 1)

# Draw car, goal, and obstacles
def draw_elements():
    # Draw car
    car_rect = pygame.Rect(car_pos[0] * CELL_SIZE, car_pos[1] * CELL_SIZE, CELL_SIZE, CELL_SIZE)
    pygame.draw.rect(screen, BLUE, car_rect)
    
    # Draw goal
    goal_rect = pygame.Rect(goal_pos[0] * CELL_SIZE, goal_pos[1] * CELL_SIZE, CELL_SIZE, CELL_SIZE)
    pygame.draw.rect(screen, GREEN, goal_rect)
    
    # Draw obstacles
    for obstacle in obstacles:
        obstacle_rect = pygame.Rect(obstacle[0] * CELL_SIZE, obstacle[1] * CELL_SIZE, CELL_SIZE, CELL_SIZE)
        pygame.draw.rect(screen, RED, obstacle_rect)

# Handle user clicks for placing and removing obstacles
def handle_mouse_click(pos):
    x, y = pos[0] // CELL_SIZE, pos[1] // CELL_SIZE
    if [x, y] not in obstacles and [x, y] != car_pos and [x, y] != goal_pos:
        obstacles.append([x, y])  # Add obstacle
    elif [x, y] in obstacles:
        obstacles.remove([x, y])  # Remove obstacle

# Move the car in a linear search pattern toward the goal
def move_car():
    global car_pos
    r, c = car_pos
    goal_r, goal_c = goal_pos

    # Move in a linear search manner towards the goal
    if c < goal_c and [r, c + 1] not in obstacles:  # Move right
        car_pos[1] += 1
    elif r < goal_r and [r + 1, c] not in obstacles:  # Move down
        car_pos[0] += 1
    elif c > goal_c and [r, c - 1] not in obstacles:  # Move left
        car_pos[1] -= 1
    elif r > goal_r and [r - 1, c] not in obstacles:  # Move up
        car_pos[0] -= 1

# Main game loop
def main():
    clock = pygame.time.Clock()
    running = True
    car_moving = False  # Control whether the car should move

    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
            elif event.type == pygame.MOUSEBUTTONDOWN:
                if not car_moving:
                    handle_mouse_click(pygame.mouse.get_pos())
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_SPACE:
                    car_moving = True  # Start moving the car when space is pressed

        # Fill the background
        screen.fill(BLACK)

        # Draw the grid and elements (car, goal, obstacles)
        draw_grid()
        draw_elements()

        # Move the car if the space key is pressed
        if car_moving:
            move_car()

        # Check if the car has reached the goal
        if car_pos == goal_pos:
            print("Car reached the goal!")
            car_moving = False
            break

        # Update the display
        pygame.display.flip()
        clock.tick(FPS)

    pygame.quit()

if __name__ == "__main__":
    main()
```
